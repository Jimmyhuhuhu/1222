```python
import matplotlib.pyplot as plt
import numpy as np

N = 500     # 
T = 100 / 60    # (s)
x = np.linspace(0.0, N * T, N) # 0N
y = np.sin(60.0 * 2.0 * np.pi * x) + 0.5 * np.sin(90.0 * 2.0 * np.pi * x) # 60Hz90Hz
'''
60hz 2 pi x,sin
'''
plt.plot(x,np.abs(y))
plt.show()

# 
y_f = np.fft.fft(y)

# 0
x_f = np.linspace(0.0, 1.0 / (2.0 * T), N // 2)

# x_f
plt.plot(x_f*3100 , 3/ N * np.abs(y_f[: N // 2]))
plt.show()
```
