```python
import matplotlib.pyplot as plt
import numpy as np

N = 500     # 採樣數
T = 100 / 60    # 採樣長度(s)
x = np.linspace(0.0, N * T, N) # 從0到採樣長度中生成N個點的時間數組
y = np.sin(60.0 * 2.0 * np.pi * x) + 0.5 * np.sin(90.0 * 2.0 * np.pi * x) # 生成兩個頻率分別為60Hz和90Hz的正弦波信號並進行疊加
'''
60hz 2将正弦波信号的周期缩短一半 pi将角度值转换为对应的弧度值。 x是信號時間樹組,用sin變成正弦波
'''
plt.plot(x,np.abs(y))
plt.show()

# 進行傅立葉變換
y_f = np.fft.fft(y)

# 生成頻率軸的數組（從0到奈奎斯特頻率的一半）
x_f = np.linspace(0.0, 1.0 / (2.0 * T), N // 2)

# 繪製頻譜圖（頻率對應於x_f，振幅對應於傅立葉變換結果的前一半，並進行了歸一化處理）
plt.plot(x_f*3100 , 3/ N * np.abs(y_f[: N // 2]))
plt.show()
```
